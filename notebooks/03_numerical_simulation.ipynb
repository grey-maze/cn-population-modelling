# Numerical Simulation of the Logistic Equation
In this notebook, we approximate the logistic differential equation using numerical methods. We compare numerical approximations with the exact closed-form solution and analyze discretization error.

We have $$\frac{dP}{dt} = rP\left(1 - \frac{P}{K}\right)$$


Solution: $$P(t) = \frac{K}{1 + Ae^{-rt}}\$$

Where $r=0.0412$, $K=1.5988$ and $A=2.1548$

Euler's method: $$\displaystyle P_{n+1} = P_n + hrP_n \left( 1 - \frac{P_n}{K} \right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
r = 0.0412444953186689
K = 1.5988153493907211
A = 2.1547780229781694

def logistic_exact(t):
    return K / (1 + A * np.exp(-r * t))

def euler_method(P0, h, T):
    N = int(T / h)
    P = np.zeros(N + 1)
    t = np.linspace(0, T, N + 1)
    P[0] = P0
    
    for n in range(N):
        P[n+1] = P[n] + h * r * P[n] * (1 - P[n]/K)
    
    return t, P
```

```python
@interact(h=(0.01, 5.0, 0.05))
def compare_plots(h=5.0):
    P0 = logistic_exact(0)
    T = 75
    
    t_euler, P_euler = euler_method(P0, h, T)
    t_exact = np.linspace(0, T, 1000)
    P_exact = logistic_exact(t_exact)
    
    error = np.abs(logistic_exact(t_euler) - P_euler)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    ax1.plot(t_exact, P_exact, 'k--', label='Exact Solution', alpha=0.7)
    ax1.plot(t_euler, P_euler, 'b-', label=f'Euler (h={h})')
    ax1.set_xlabel("Time")
    ax1.set_ylabel("Population (billions)")
    ax1.set_title("Exact Solution vs Euler Approximation")
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    ax2.plot(t_euler, error)
    ax2.set_xlabel("Time")
    ax2.set_ylabel("Absolute Error")
    ax2.set_title("Euler Method Error Over Time")
    ax2.set_ylim(0, 0.015)
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```

**INTERPRETATION**


While the logistic equation admits a closed-form solution, numerical schemes such as Euler’s method provide a general framework for approximating nonlinear dynamical systems when analytical solutions are unavailable. The observed reduction in error with decreasing step size confirms the first-order convergence of the scheme. However, large time steps can introduce numerical instability, highlighting the importance of discretization choices in computational modeling.
