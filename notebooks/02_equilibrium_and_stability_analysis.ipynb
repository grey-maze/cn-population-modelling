# Dynamical Systems Analysis of the Logistic Equation
In this notebook, we analyze the logistic differential equation from a dynamical systems perspective. We identify equilibrium points, determine their stability, and interpret their demographic meaning.

We have $$\frac{dP}{dt} = rP\left(1 - \frac{P}{K}\right)$$

Where:

- $P\left(t\right)$ : population (billions)
- $r > 0$ : intrinsic growth rate  
- $K > 0$ : carrying capacity

We treat this as a one-dimensional autonomous dynamical system.

Setting $$\frac{dP}{dt} = 0 $$
Then $$rP\left(1 - \frac{P}{K}\right) = 0$$
Giving $P = 0$ and $K = 0$

The system has two equilibrium points: extinction and saturation.

Let $$f(P) = rP\left(1 - \frac{P}{K}\right)$$
Then $$f'(P) = r\left(1 - \frac{2P}{K}\right)$$
At  $P = 0$,  $f'(0) = r > 0$, small growth, hence unstable

At  $P = K$,  $f'(0) = -r < 0$, small decay, hence asymptotically stable

```python
import numpy as np
import matplotlib.pyplot as plt

r = 0.041
K = 1.6

P_vals = np.linspace(0, 2.0, 400)
dP = r * P_vals * (1 - P_vals / K)

plt.figure()
plt.plot(P_vals, dP)
plt.axhline(0)
plt.axvline(0)
plt.axvline(K)
plt.xlabel("P")
plt.ylabel("dP/dt")
plt.title("Phase Line of Logistic Equation")
plt.show()
```

### DISCUSSION
The logistical equation predicts that any positive initial population converges monotonically to the carrying capacity $K$. The equilibrium at $P=K$ is globally asymptotically stable for $P > 0$. Thus, under classical logistic assumptions, long-term population decline is not structurally stable; the system inherently pushes trajectories toward saturation.
